```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_json(path_or_buf='imoveis.json',orient='columns') #leitura do arquivo json
```

```python
dados_lista1 = pd.json_normalize(dados.ident) # tranforma json em df
```

```python
dados_lista1.head()
```

```python
dados_lista2 = pd.json_normalize(dados.listing)
```

```python
dados_lista2.head()
```

```python
dados_imoveis = pd.concat([dados_lista1,dados_lista2],axis=1)
```

```python
dados_imoveis.head()
```

```python
dados_imoveis.shape # retorna as dimensoes do dataframe
```

```python
for coluna in dados_imoveis.columns:
    print('----'*10)
    print(dados_imoveis[coluna].value_counts())
```

```python
#Pré processamento dos dados
#Aplicando filtros no dataset
filtro = (dados_imoveis['types.usage'] == 'Residencial')&(dados_imoveis['address.city'] == 'Rio de Janeiro')
```

```python
dados_imoveis= dados_imoveis[filtro]
```

```python
dados_imoveis.head()
```

```python
dados_imoveis.info() # verifica as informações dos dados
```

```python
# reset do index 
dados_imoveis.reset_index(drop=True, inplace=True)
```

```python
dados_imoveis.info(verbose= False)
```

```python
#transformação dos tipos de dados
dados_imoveis = dados_imoveis.astype({
    'prices.price':'float64',
    'prices.tax.iptu':'float64',
    'prices.tax.condo':'float64',
    'features.usableAreas':'int64',
    'features.totalAreas':'int64'
})
```

```python
dados_imoveis.info()
```

```python
#Lidando com dados nulos
dados_imoveis['address.zone'].value_counts()
```

```python
dados_imoveis['address.zone'] = dados_imoveis['address.zone'].replace('',np.nan) #substitui o espaço vazio por nan
```

```python
dados_imoveis['address.zone'].value_counts()
```

```python
dados_imoveis['address.zone'].isnull().value_counts()
```

```python
dados_imoveis.info()
```

```python
dados_imoveis.head()
```

```python
dici = dados_imoveis[~dados_imoveis['address.zone'].isna()].drop_duplicates(subset=['address.neighborhood']).to_dict('records') # remove valores duplicados
print(dici)
```

```python
dados_imoveis['address.zone'].isnull().sum()
```

```python
dic_zonas = {dic['address.neighborhood']:dic['address.zone']for dic in dici} #criando dicionario para associar
```

```python
print(dic_zonas)
```

```python
#Associando o bairro com a zona
for bairro, zona in dic_zonas.items():
    dados_imoveis.loc[dados_imoveis['address.neighborhood']==bairro,'address.zone']=zona
```

```python
dados_imoveis.head()
```

```python
dados_imoveis['prices.tax.condo'].isnull().sum()
```

```python
dados_imoveis['prices.tax.iptu'].isnull().sum()
```

```python
dados_imoveis['prices.tax.condo'].fillna(0,inplace = True)
dados_imoveis['prices.tax.iptu'].fillna(0,inplace = True)
```

```python
print(f"Total de valores nulos tax condo:{dados_imoveis['prices.tax.condo'].isnull().sum()}")
print(f"Total de valores nulos tax iptu:{dados_imoveis['prices.tax.iptu'].isnull().sum()}")
```

```python
#Remove colunas nao uteis para criar o modelo
dados_imoveis.drop(['customerID','source','types.usage','address.city',
                    'address.location.lon','address.location.lat','address.neighborhood'],axis=1, inplace= True)
```
